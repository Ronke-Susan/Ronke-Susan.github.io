###  INTRODUCTION

This project is targetted at dog lovers who are looking to buy or adopt a dog but are not sure what breed to go for.

There are 300+ breeds of dog recognized by various canine registries of the world such as American Kennel Club (AKC) , Britain's Kennel Club (KC), the Federation Cynologique International and Europe's Continental Kennel Club (CKC).
These breeds are recognized based on factors such as popularity, population and performance at dog shows.

### DATA INFORMATION

The data used for this project was scraped from [link]('https://www.dogbreedslist.info/') using BeautifulSoup python library.

It contains information needed for analysis. This include, Dog breed, Life span, Puppy price, Intelligence rating, popularity	,hypoallergenic, characteristics and url links for each dog

353 dog breeds were used for the analysis.
